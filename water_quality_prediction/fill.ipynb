```python
import pandas as pd
```

```python
file_name = './data/expand_mini.xlsx'
df = pd.read_excel(file_name) # padas
data = df
data
```

```python
fill_cnt = 2
mask = data.copy()
for i in data.columns: 
    df = pd.DataFrame( data[i] )
    df['new'] = ((df.notnull() != df.shift().notnull()).cumsum())
    df['ones'] = 1
    mask[i] = (df.groupby('new')['ones'].transform('count') < fill_cnt+1) | data[i].notnull()
mask
```

```python
data.interpolate().bfill()[mask]
```
