```python
import matplotlib.pyplot as plt
```

```python
f = open("./data/MSE_loss_curr.txt", 'r')
idx = 0
idx_list = []
data_list = []
while True:
    line = f.readline()
    if not line: break
    # print(line)
    # print('x -> ', idx%2)
    if idx%2 == 0:
        idx_list.append(line.strip())
    else:
        data_list.append(line.strip())
    idx += 1
    # if idx == 100:
    #     break
f.close()
```

```python
plt.plot(idx_list, data_list)
```
