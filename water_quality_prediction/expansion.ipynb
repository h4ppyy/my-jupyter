```python
import pandas as pd
```

```python
target_col = [0,1,2,3]
target_col
```

```python
df = pd.read_excel('./data/expand_mini.xlsx')
```

```python
df
```

```python
origin_df = df.iloc[:, target_col]
origin_df
```

```python
df.iloc[:, target_col].isna()
```

```python
df.iloc[:, target_col].isna().any(axis='columns')
```

```python
x = df.iloc[:, target_col].isna().any(axis='columns').astype(int)
x
```

```python
y = df.iloc[:, target_col].isna().any(axis='columns').astype(int).shift(periods=1, fill_value=0)
y
```

```python
pd.concat([x, y], axis=1)
```

```python
pd.concat([x, y], axis=1).any(axis='columns').astype(int)
```

```python
z = pd.concat([x, y], axis=1).any(axis='columns').astype(int).cumsum()
z
```

```python
label_df = pd.concat([x, z], axis=1)
label_df.columns = ['is_nan', 'group']
label_df
```

```python
new_df = pd.concat([origin_df, label_df], axis=1)
new_df
```

```python
result_df = new_df.where(new_df['is_nan'] != 1).dropna()
result_df
```

```python
size_sr = result_df.groupby(result_df['group']).size()
size_sr
```

```python
result_df.where(result_df['group'] == 0).dropna()
```

```python
result_df.where(result_df['group'] == 2).dropna()
```

```python
result_df.where(result_df['group'] == 4).dropna()
```

```python
input_unit_time = 5
input_filter_size = 2
```

```python
size_sr.where(size_sr >= input_unit_time).dropna()
```

```python
target_idx = size_sr.where(size_sr >= input_unit_time).dropna().index[0]
target_idx
```

```python
result_df.where(result_df['group'] == target_idx).dropna()[:input_unit_time]
```

```python
total_cnt = len(result_df.where(result_df['group'] == target_idx).dropna()[:input_unit_time])
total_cnt
```

```python
concat_list = []
for n in range(0, total_cnt-input_filter_size+1):
    split_df = result_df.where(result_df['group'] == target_idx).dropna()[:input_unit_time][n:input_filter_size+n]
    concat_list.append(split_df)
    print(split_df)
```

```python
output_df = pd.concat(concat_list).drop('is_nan', axis=1).drop('group', axis=1)
output_df
```

```python
output_df.to_excel("output/output.xlsx", index=False)
```
